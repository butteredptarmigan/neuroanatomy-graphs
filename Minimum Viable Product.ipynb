```python
from pyvis.network import Network
import ipywidgets as widgets
```

```python
net = Network(directed=True, notebook=True)
```

```python
node_name = widgets.Text(placeholder='nazwa')
node_abbr = widgets.Text(placeholder='skrót')
button = widgets.Button(description='Dodaj')

def add_structure(b):
    name = node_name.value
    abbr = node_abbr.value
    
    net.add_node(abbr, label=abbr, title=name)
    
    node_name.value = ''
    node_abbr.value = ''

button.on_click(add_structure)

node_box = widgets.Box(children=[
    widgets.Label('Struktura'),
    node_name,
    node_abbr,
    button
])
node_box
```

```python
# Add a few nodes programatically for the sake of presentation
new_nodes = [
        {'title': 'piriform cortex', 'id': 'PC', 'label': 'PC'},
        {'title': 'olfactory bulb', 'id': 'OB', 'label': 'OB'},
        {'title': 'ventral tegmental area', 'id': 'VTA', 'label': 'VTA'},
        {'title': 'nucleus accumbens', 'id': 'NAc', 'label': 'NAc'},
        {'title': 'ventral pallidum', 'id': 'VP', 'label': 'VP'},
        {'title': 'paraventricular thalamus', 'id': 'PVT', 'label': 'PVT'},
        {'title': 'olfactory tubercle', 'id': 'OT', 'label': 'OT'},
        {'title': 'anterior olfactory nucleus', 'id': 'AON', 'label': 'AON'},
        {'title': 'mediodorsal thalamus', 'id': 'MDT', 'label': 'MDT'}
]
for node in new_nodes:
   net.add_node(node['id'], label=node['label'], title=node['title'])
```

```python
net.show('networks/MVP_nodes.html')
```

```python
fro = widgets.Text(placeholder='A')
to = widgets.Text(placeholder='B')
doi = widgets.Text(placeholder='DOI')
sentence = widgets.Text(placeholder='Zdanie źródłowe')
```

```python
button = widgets.Button(description='Połącz')

def add_edge(b):
    net.add_edge(
        fro.value,
        to.value,
        title=(doi.value + '\n' + sentence.value))
    
    fro.value = ''
    to.value = ''
    sentence.value = ''

button.on_click(add_edge)
```

```python
nodes = [fro, widgets.Label(value=' → '), to]
nodes_layout = widgets.Layout(display='flex',
                            flex_flow='row',
                            align_items='stretch')
nodes_box = widgets.Box(children=nodes, layout=nodes_layout)

items = [nodes_box, doi, sentence, button]
box_layout = widgets.Layout(display='flex',
                    flex_flow='column', 
                    align_items='stretch',
                    width='600px',
                    overflow='hidden')

box = widgets.Box(children=items, layout=box_layout)
box
```

```python
# Add a few edges
new_edges = [
    {'from': 'OT', 'to': 'VP', 'doi': 'https://doi.org/10.3389/fncir.2024.1423505', 'sentence': 'the major output from the OT is considered to the ventral pallidum (VP)'},
    {'from': 'OT', 'to': 'AON', 'doi': 'https://doi.org/10.3389/fncir.2024.1423505', 'sentence': 'Many recent studies, including those employing transsynaptic tracing, have revealed various projection targets of the OT, including the [...] anterior olfactory nucleus'},
    {'from': 'OT', 'to': 'PC', 'doi': 'https://doi.org/10.3389/fncir.2024.1423505', 'sentence': 'Many recent studies, including those employing transsynaptic tracing, have revealed various projection targets of the OT, including the PC'},
    {'from': 'OT', 'to': 'MDT', 'doi': 'https://doi.org/10.3389/fncir.2024.1423505', 'sentence': 'Many recent studies, including those employing transsynaptic tracing, have revealed various projection targets of the OT, including the [...] mediodorsal thalamus'},
    {'from': 'OT', 'to': 'VTA', 'doi': 'https://doi.org/10.3389/fncir.2024.1423505', 'sentence': 'Many recent studies, including those employing transsynaptic tracing, have revealed various projection targets of the OT, including the [...] VTA'}
]
for edge in new_edges:
    net.add_edge(edge['from'], edge['to'], title=(edge['doi'] + '\n' + edge['sentence']))
```

```python
net.show('networks/MVP_network.html')
```
